```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from dataclasses import dataclass
dataset=pd.read_csv('loan-predictionUC.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
dataset=dataset.replace({'Gender':{'Male':0,'Female':1}})
dataset=dataset.replace({'Married':{'Yes':1,'No':0}})
dataset=dataset.replace({'Dependents':{'3+':'3'}})
dataset=dataset.replace({'Education':{'Not Graduate':0,'Graduate':1}})
dataset=dataset.replace({'Self_Employed' : { 'No' : 0, 'Yes' : 1}})
dataset=dataset.replace({'Property_Area' : { 'Urban' : 1, 'Rural' : 2,'Semiurban':3,'semiurban':4,'Semi-urban':5}})
dataset=dataset.replace({'Loan_Status':{'Y':1,'N':0}})
```

```python
dataset
```

```python
dataset = dataset.replace(np.nan,0)
```

```python
dataset['Dependents']=dataset['Dependents'].astype(int)
```

```python
x=dataset[[ 'Gender', 'Married', 'Dependents', 'Education',
 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
 'Loan_Amount_Term', 'Credit_History', 'Property_Area']]
```

```python
y=dataset['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y ,test_size=0.2, random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
